```python
import pandas as pd

perf = pd.read_csv(r"C:\Users\lenovo\OneDrive\Desktop\Data_Analyst_Intern\bluestock_mf_capstone\Data\Raw\07_scheme_performance.csv")

risk = input(
    "Enter Risk Appetite (Low/Moderate/High): "
)

recommended = (
    perf[
        perf["risk_grade"] == risk
    ]
    .sort_values(
        "sharpe_ratio",
        ascending=False
    )
    .head(3)
)

print("\nTop Recommended Funds")

print(
    recommended[
        [
            "scheme_name",
            "risk_grade",
            "sharpe_ratio"
        ]
    ]
)
```
