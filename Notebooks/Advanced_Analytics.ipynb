```python
# Convert dates
nav["date"] = pd.to_datetime(nav["date"])
txn["transaction_date"] = pd.to_datetime(txn["transaction_date"])
```

```python
# ==========================================================
# TASK 1 : HISTORICAL VaR & CVaR (95%)
# ==========================================================

# Daily Return Calculation
nav = nav.sort_values(["amfi_code", "date"])

nav["daily_return"] = (
    nav.groupby("amfi_code")["nav"]
       .pct_change()
)

var_results = []

for fund in nav["amfi_code"].unique():

    returns = nav[
        nav["amfi_code"] == fund
    ]["daily_return"].dropna()

    if len(returns) > 0:

        # 95% Historical VaR
        var95 = np.percentile(returns, 5)

        # CVaR
        cvar95 = returns[
            returns <= var95
        ].mean()

        var_results.append([
            fund,
            var95,
            cvar95
        ])

var_cvar_report = pd.DataFrame(
    var_results,
    columns=[
        "amfi_code",
        "VaR_95",
        "CVaR_95"
    ]
)

print("\nVaR & CVaR Report")
print(var_cvar_report.head())

# Export CSV
var_cvar_report.to_csv(
    "var_cvar_report.csv",
    index=False
)
```

```python
# ==========================================================
# TASK 2 : ROLLING 90 DAY SHARPE RATIO
# ==========================================================

top_funds = nav["amfi_code"].unique()[:5]

plt.figure(figsize=(12,6))

for fund in top_funds:

    temp = nav[
        nav["amfi_code"] == fund
    ].copy()

    returns = temp["daily_return"]

    sharpe = (
        returns.rolling(90).mean()
        /
        returns.rolling(90).std()
    ) * np.sqrt(252)

    plt.plot(
        temp["date"],
        sharpe,
        label=str(fund)
    )

plt.title("Rolling 90 Day Sharpe Ratio")
plt.xlabel("Date")
plt.ylabel("Sharpe Ratio")
plt.legend()

plt.savefig(
    "rolling_sharpe_chart.png"
)

plt.show()
```

```python
# ==========================================================
# TASK 3 : INVESTOR COHORT ANALYSIS
# ==========================================================

# First Investment Year
txn["cohort_year"] = (
    txn.groupby("investor_id")
       ["transaction_date"]
       .transform("min")
       .dt.year
)

cohort_metrics = txn.groupby(
    "cohort_year"
).agg(
    avg_sip_amount=("amount_inr", "mean"),
    total_invested=("amount_inr", "sum")
)

print("\nCohort Analysis")
print(cohort_metrics)

# Top Fund Preference
top_fund_pref = (
    txn.groupby(
        ["cohort_year", "amfi_code"]
    )["amount"]
    .sum()
    .reset_index()
    .sort_values(
        "amount_inr",
        ascending=False
    )
    .drop_duplicates(
        "cohort_year"
    )
)

print("\nTop Fund Preference")
print(top_fund_pref)
```

```python
# ==========================================================
# TASK 4 : SIP CONTINUITY ANALYSIS
# ==========================================================

sip = txn[
    txn["transaction_type"] == "SIP"
].copy()

# Investors having 6+ SIPs
eligible = (
    sip.groupby("investor_id")
       .size()
)

eligible = eligible[
    eligible >= 6
].index

sip = sip[
    sip["investor_id"].isin(
        eligible
    )
]

sip = sip.sort_values(
    ["investor_id", "transaction_date"]
)

sip["gap_days"] = (
    sip.groupby("investor_id")
       ["transaction_date"]
       .diff()
       .dt.days
)

continuity = (
    sip.groupby("investor_id")
       ["gap_days"]
       .mean()
       .reset_index()
)

continuity["status"] = np.where(
    continuity["gap_days"] > 35,
    "At Risk",
    "Active"
)

print("\nSIP Continuity")
print(continuity.head())
```

```python
# ==========================================================
# TASK 5 : SIMPLE FUND RECOMMENDER
# ==========================================================

risk_appetite = "Moderate"

recommendations = (
    perf[
        perf["risk_grade"] == risk_appetite
    ]
    .sort_values(
        "sharpe_ratio",
        ascending=False
    )
    .head(3)
)

print("\nRecommended Funds")
print(
    recommendations[
        [
            "scheme_name",
            "risk_grade",
            "sharpe_ratio"
        ]
    ]
)
```

```python
# ==========================================================
# TASK 6 : SECTOR HHI CONCENTRATION
# ==========================================================

hhi = holdings.groupby(
    "amfi_code"
).apply(
    lambda x:
    np.sum(
        (x["weight_pct"] / 100) ** 2
    )
)

hhi = hhi.reset_index()

hhi.columns = [
    "amfi_code",
    "HHI"
]

print("\nHHI Concentration")
print(hhi.head())
```
