```python
# Fix date formats
df2['transaction_date'] = pd.to_datetime(
    df2['transaction_date'],
    errors='coerce'
)
```

```python
# Standardise transaction_type values
df2['transaction_type'] = (
    df2['transaction_type']
    .astype(str)
    .str.strip()
)
```

```python
df2['transaction_type'] = df2['transaction_type'].replace({
    'sip': 'SIP',
    'SIP': 'SIP',
    'lumpsum': 'Lumpsum',
    'Lumpsum': 'Lumpsum',
    'redemption': 'Redemption',
    'Redemption': 'Redemption'
})
print(df2['transaction_type'])
```

```python
# Validate amount > 0
df2 = df2[df2['amount_inr'] > 0]
```

```python
# Check KYC status enum values
valid_kyc = ['Verified', 'Pending', 'Rejected']

df2 = df2[df2['kyc_status'].isin(valid_kyc)]
```

```python
# Save cleaned dataset
df2.to_csv(
    r"C:\Users\lenovo\OneDrive\Desktop\Data_Analyst_Intern\bluestock_mf_capstone\Data\processed\investor_transactions_clean.csv",
    index=False
)

print("investor_transactions_clean.csv saved successfully")
print("Rows:", len(df2))
```

```python
df3=pd.read_csv(r"C:\Users\lenovo\OneDrive\Desktop\Data_Analyst_Intern\bluestock_mf_capstone\Data\Raw\07_scheme_performance.csv")
df3
```

```python
print("Shape:", df3.shape)
print("\nColumns:")
print(df3.columns.tolist())
print("\nData Types:")
print(df3.dtypes)
```

```python
# Validate return values are numeric
return_cols = [
    'return_1yr_pct',
    'return_3yr_pct',
    'return_5yr_pct'
]

for col in return_cols:
    df3[col] = pd.to_numeric(df3[col], errors='coerce')
```

```python
# Flag anomalies
df3['anomaly_flag'] = np.where(
    (df3['return_1yr_pct'] > 100) |
    (df3['return_1yr_pct'] < -50),
    1,
    0
)
```

```python
# Check expense ratio range (0.1% - 2.5%)
df3 = df3[
    (df3['expense_ratio_pct'] >= 0.1) &
    (df3['expense_ratio_pct'] <= 2.5)
]
```

```python
# Save cleaned dataset
df3.to_csv(
    r"C:\Users\lenovo\OneDrive\Desktop\Data_Analyst_Intern\bluestock_mf_capstone\Data\processed\scheme_performance_clean.csv",
    index=False
)
```

```python
print("scheme_performance_clean.csv saved successfully")
print("Rows:", len(df3))
```
